```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [115,85,90,150,149,88,95,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g',)
```

```python
blood_sugar_men = [115,85,90,150,149,88,95,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('Sugar Range')
plt.ylabel('Total No. Of Patients')
plt.title('Bood-Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['Men','Women'],orientation='horizontal')
plt.legend()
```
